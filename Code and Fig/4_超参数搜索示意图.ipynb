```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as mtri
```

```python
df_result = pd.read_csv("网格搜索的结果——0.8.csv")
df_result
```

# 根据SIze放缩结果

```python
a = df_result.iloc[:, -2:-1]                 # a 是 size
from sklearn import preprocessing
#创建转换器
minmax_scale = preprocessing.MinMaxScaler()  # feature_range=(0,1)
#缩放特征
minmax_size = minmax_scale.fit_transform(a)
df_result['minmax_size'] = minmax_size 
df_result
```

# 根据size给label

```python
a = df_result.iloc[:, -2:-1]                 # a 是 size
from sklearn import preprocessing
#创建转换器
minmax_scale = preprocessing.MinMaxScaler(feature_range=(0,100))  # 
#缩放特征
minmax_size = minmax_scale.fit_transform(a)
df_result['minmax_size'] = minmax_size 
df_result
```

```python
df_result['size'].iloc[1]
```

```python
df_result['label'] = 1
for i in range(0, 4800):
    if df_result['size'].iloc[i] >= 500:
        tmp = "good"
    elif df_result['size'].iloc[i] >= 400:
        tmp = "well"
    elif df_result['size'].iloc[i] >= 300:
        tmp = "mid"
    else:  # df['size'] >= 200
        tmp = "bad"
    df_result["label"].iloc[i] = tmp
df_result
```

```python
x = df_result["n_estimators"]
y = df_result["max_depth"]
z = df_result["min_samples_split"]
value = df_result["mean_test_score"]
```

```python
import plotly.graph_objects as go
import numpy as np
X, Y, Z = np.mgrid[-8:8:40j, -8:8:40j, -8:8:40j]
values = np.sin(X*Y*Z) / (X*Y*Z)

fig = go.Figure(data=go.Volume(
    x=x,
    y=y,
    z=z,
    value=value,
    isomin=0.8,
    isomax=0.85,
#     opacity=0.1, # needs to be small to see through all surfaces
#     surface_count=17, # needs to be a large number for good volume rendering
    ))
fig.show()
```
